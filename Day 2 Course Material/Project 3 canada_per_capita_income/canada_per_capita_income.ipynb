# Canada Per Capita Income Prediction (Linear Regression)

Predict Canada's per capita income in year 2020. Using the dataset `canada_per_capita_income.csv`, build a regression model and predict the per capita income for Canadian citizens in year 2020.

**Target Answer:** `41288.69409442`

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Load the dataset
df = pd.read_csv('canada_per_capita_income.csv')
df.head()
```

```python
# Plot scatter plot of year vs per capita income
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.scatter(df['year'], df['per capita income (US$)'], color='red', marker='+')
plt.title('Canada Per Capita Income Trend')
plt.show()
```

```python
# Fit linear regression model
reg = LinearRegression()
reg.fit(df[['year']], df['per capita income (US$)'])
```

```python
# Predict for 2020
prediction_2020 = reg.predict([[2020]])[0]
print(f"Predicted Per Capita Income for Canadian citizens in 2020: US$ {prediction_2020:.8f}")

expected = 41288.69409442
print(f"Target expected: {expected}")
print(f"Difference: {abs(prediction_2020 - expected):.8e}")
```

```python
# Plot the line of best fit
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.scatter(df['year'], df['per capita income (US$)'], color='red', marker='+')
plt.plot(df['year'], reg.predict(df[['year']]), color='blue')
plt.title('Canada Per Capita Income Linear Regression')
plt.show()
```
